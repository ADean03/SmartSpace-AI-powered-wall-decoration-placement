```python
import json

# Path to the info.json file
info_json_path = '/content/drive/MyDrive/SmartSpace/MYSUN/info.json'

# Load the JSON file
with open(info_json_path, 'r') as f:
    data = json.load(f)

# Print the first entry to understand its structure
#print(data[0])

unique_classes = set()

# Iterate over each entry in the JSON file
for entry in data:
    # Iterate over each annotation in the 'annotations' list
    for annotation in entry['annotations']:
        classname = annotation['classname']
        unique_classes.add(classname)  # Add classname to the set

# Convert the set to a sorted list if you want it ordered
unique_classes_list = sorted(list(unique_classes))

# Print the unique class names
for i, classname in enumerate(unique_classes_list, 1):
    print(f"{i}. {classname}")
```
